```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Placeholder data (replace these values with your actual data)
countries = [
    'Korea, Rep.', 'Iraq', 'Japan', 'China', 'Pakistan', 'United States', 'India'
]

# GDP data in trillions
gdp_data = {
    'Korea, Rep.': [1.2, 1.4, 1.8, 2.1, 2.5, 2.9, 3.3, 3.7, 4.1, 4.5],
    'Iraq': [0.5, 0.7, 0.9, 1.1, 1.3, 1.5, 1.7, 2.0, 2.2, 2.5],
    'Japan': [4.5, 4.8, 5.0, 5.2, 5.5, 5.8, 6.0, 6.2, 6.5, 6.8],
    'China': [8.0, 8.5, 9.0, 9.5, 10.0, 10.5, 11.0, 11.5, 12.0, 12.5],
    'Pakistan': [0.3, 0.5, 0.7, 1.0, 1.3, 1.6, 1.9, 2.2, 2.5, 2.8],
    'United States': [15.0, 16.0, 17.0, 18.0, 19.0, 20.0, 21.0, 22.0, 23.0, 24.0],
    'India': [2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5]
}

years = np.arange(2013, 2023)

# Convert data to long format
df = pd.DataFrame(gdp_data)
df['Year'] = years
df_long = pd.melt(df, id_vars=['Year'], var_name='Country', value_name='GDP')

# Set seaborn style
sns.set(style="whitegrid")

# Define a custom color palette
custom_palette = sns.color_palette("husl", len(countries))

# Bar Plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df_long, x='Year', y='GDP', hue='Country', palette=custom_palette)
ax.set_title('Military Expenditures (2013-2022)')
ax.set_xlabel('Year')
ax.set_ylabel('Expenditures (in Trillions USD)')
plt.legend(title='Country', loc='upper left')
plt.show()
```

```python
# Line Plot
plt.figure(figsize=(12, 6))
ax = sns.lineplot(data=df_long, x='Year', y='GDP', hue='Country', marker='o', palette=custom_palette)
ax.set_title('Military Expenditures (2013-2022)')
ax.set_xlabel('Year')
ax.set_ylabel('Expenditures (in Trillions USD)')
plt.legend(title='Country', loc='upper left')
plt.show()
```
